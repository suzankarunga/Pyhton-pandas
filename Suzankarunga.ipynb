                  Bivarient Analysis 
This is for  annalyzing  two variables  to understand their annalysis .


```python
sns.scatterplot(data=tin , x='Age', y='Education')
plt.title("Age vs. Education")
plt.show()
```

 
This code allows for visualizing any potential correlation or patterns between age and education within the dataset.

```python
corr_age_education = tin["Age"].corr(tin["Education"])
print("Correlation coefficient between Age and Education:", corr_age_education)
```

 THis code calculates the relationship between two varriables in your data set. this also shows the coleration between each varriable in the dataset. 

                    Multivarrient 
 This refers to the analysis of data involving multiple variables simultaneously.

```python
tin.corr()
```

This code  calculates the correlation coefficient between varriable in the DataFrame 

```python

correlation_matrix = tin.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 10})
plt.title("Correlation Matrix")
plt.show()
```

 this code generates a correlation matrix showing the correlation coefficients between all pairs of variables. 

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(tin['Age'], tin['Education'], tin['Income'], c='b', marker='o')

# Set labels and title
ax.set_xlabel('Income')
ax.set_ylabel('Wealth')
ax.set_zlabel('Home value')
ax.set_title('3D Scatter Plot of Income , Wealth, and Home value')

plt.show()
```

This code generates a 3D scatter plot to visualize the relationship between three variables:  Income , Wealth, and Home value.
